Zbiór danych Pima Indians Diabetes zawiera informacje na temat 768 pacjentek należących do plemienia Indian Pima z Arizony. Pacjentki te zostały poddane badaniom diagnostycznym cukrzycy w latach 1988-1989. Zbiór danych składa się z 8 zmiennych numerycznych oraz jednej zmiennej binarnej, która określa, czy pacjentka choruje na cukrzycę (1) czy nie (0).

Poniżej znajdują się opisy poszczególnych zmiennych:

 - Pregnancies: liczba ciąż, jakie pacjentka przeszła.
 - Glucose: stężenie glukozy we krwi na czczo.
 - BloodPressure: skurczowe ciśnienie krwi w mm Hg.
 - SkinThickness: grubość fałdu skórnego tricepsu w mm.
 - Insulin: poziom insuliny we krwi w mU/ml.
 - BMI: wskaźnik masy ciała (Body Mass Index) - waga w kilogramach podzielona przez wzrost w metrach do kwadratu.
 - DiabetesPedigreeFunction: prawdopodobieństwo wystąpienia cukrzycy na podstawie histori rodzinnej
 - Age: wiek pacjentki.
 - Outcome: czy pacjentka choruje na cukrzycę (1) czy nie (0) - zmienna celu.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

diabetes = pd.read_csv("data/diabetes.csv")
diabetes
```

```python
counts = diabetes['Outcome'].value_counts()
percentages = counts / counts.sum() * 100

plt.bar(percentages.index, percentages.values)
plt.title('Percentage of Outcome Column')
plt.xlabel('X')
plt.ylabel('Percentage')
plt.show()
```

```python
def generate_plots_and_tables_by_tagret(df: pd.DataFrame, target_column: str = "Outcome",
                                        numeric_cols: list = [], categorical_cols: list = []) -> None:
    """
    Generate plots of the 'target_column' column with other columns in a pandas dataframe.
    If a column is a string or factor, a bar plot with grouping by 'target_column' will be created.
    If a column is continuous, a histogram, boxplot, and scatter plot with grouping by 'target_column' will be created.
    """
    for column in categorical_cols:
        print(column + ":")
        
        plt.figure(figsize=(8, 6))
            
        counts = df.groupby([column, target_column], group_keys=True).size()
        print(counts)
        counts_norm = counts.groupby(level=0, group_keys=False).apply(lambda x: 100 * x / x.sum())
        counts_norm = counts_norm.reset_index(name='percent')
        sns.barplot(x=column, y='percent', hue=target_column, data=counts_norm)
        plt.title(f"'Outcome' vs '{column}'")
        plt.show()
            
    for column in numeric_cols:
        
        plt.figure(figsize=(18, 6))
            
        plt.subplot(1, 2, 1)
        sns.histplot(x=column, hue=target_column, data=df, kde=True)
        plt.title(f"'Outcome' vs '{column}' (Histogram)")
            
        plt.subplot(1, 2, 2)
        sns.boxplot(x=target_column, y=column, data=df)
        plt.title(f"'Outcome' vs '{column}' (Boxplot)")
            
        plt.tight_layout()
        plt.show()
```

```python
diabetes.columns
```

```python
num_cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
               'BMI', 'DiabetesPedigreeFunction', 'Age']
cat_cols = []
generate_plots_and_tables_by_tagret(diabetes, 'Outcome', num_cols, cat_cols)
```

```python
corr_matrix = diabetes[num_cols].corr()
plt.figure(figsize=(15, 15))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
from imblearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import RepeatedStratifiedKFold, GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.decomposition import PCA

X_train, X_test, y_train, y_test = train_test_split(diabetes.drop(columns=["Outcome"]),
                                                    diabetes["Outcome"], test_size=0.2, random_state=42)
```

```python
transformation_steps = [
    ('n', MinMaxScaler(), num_cols)
]
ct = ColumnTransformer(transformation_steps)
pipeline = Pipeline(
    steps=[
        ('transform', ct),
        ('pca', PCA(n_components=4)),
        ('smote', SMOTE()),
        ('model', GradientBoostingClassifier(random_state=222))
    ])
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=0)
param_grid = {
    "model__n_estimators": [30, 50, 100],
    "model__max_depth": [3, 5, 8]
}
grid_search = GridSearchCV(
    estimator=pipeline, param_grid=param_grid, cv=cv, scoring="f1_macro"
)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
best_estimator = grid_search.best_estimator_
```

```python
probs = best_estimator.predict_proba(X_test)[:,1]
pred_df = pd.DataFrame({'Outcome': y_test, 'Probability': probs})
sns.histplot(data=pred_df, x='Probability', hue='Outcome', element='step', stat='density', bins = 50)
```

```python
from sklearn.metrics import f1_score
thresholds = np.linspace(0, 1, num=100)
f1_scores = [f1_score(pred_df.Outcome, (pred_df.Probability >= threshold).astype(int)) for threshold in thresholds]
best_threshold = thresholds[np.argmax(f1_scores)]
print("Najlepszy threshold:", best_threshold)
pred_df["pred_cutoff_max_f1"] = pred_df.Probability > 0.5
pivot_table = pred_df.drop(columns=["Probability"]).pivot_table(index='pred_cutoff_max_f1', columns='Outcome', aggfunc=len, fill_value=0)
print(pivot_table)
```

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(pred_df.Outcome, pred_df.Probability)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```
