Po zabiegu operacyjnym nastpuje ocena histopatologiczna usunitego preparatu. W zalenoci od patologicznego stopnia zaawansowania oraz stanu pacjenta wybierana jest strategia leczenia uzupeniajcego. Leczenie to moemy podzieli na rne typy:

## Model regresji logistycznej

Badana przez nas zmienna $Diagnosis$ przyjmuje dwie wartoci kategoryczne, czyli mamy do czynienia z rozkadem Bernulliego, ktry naley do **rodziny rozkadw wykadnicznych z nadmiern dyspersj** (OEF - overdispersed exponential family). Zdefiniujmy wic model **regresji logistycznej** bdcy modelem klasy GLM:

$$
y|X \sim \operatorname{Bernoulli}(p) \\
logit(\mu)= logit(p)=\ln (\frac{p}{1-p})=X\beta
$$
Odwracajc funkcje **logit** dostaniemy wzr na prawdopoddobiestwo $p$ zadane funkcj **sigmoid**:

$$
p = sigmoid(X\beta) = \frac{1}{1+e^{-X\beta}}
$$


```python
def logit(y):
    return np.log(y / (1 - y))

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-10, 10, 0.1)
s = sigmoid(x)
y = np.arange(0, 1, 0.005)
l = logit(y)

data = pd.DataFrame({'x': x, 's': s, 'y': y, 'l': l})

plt.figure(figsize=(8,6))
plt.plot(y, l)
plt.title('Logit function')
plt.xlabel('y')
plt.ylabel('logit(y)')
plt.show()

plt.figure(figsize=(8,6))
plt.plot(x, s)
plt.title('Sigmoid function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.show()
```

Funkcj masy prawdopodobiestwa dla rozkadu Bernulliego moma zapisa jako:

$$
\begin{array}{c}
f_p(y)=p^{y}(1-p)^{1-y}=\exp \left(\ln \left(p^{y}(1-p)^{1-y}\right)\right)=\exp \left(\ln \left(p^{y}\right)+\ln (1-p)^{1-y}\right)= \\
\exp (y \ln (p)+(1-y) \ln (1-p))=\exp \left(y \ln \frac{p}{1-p}+\ln (1-p)\right)
\end{array}
$$

gdzie:

$$
\begin{array}{c}
\theta=\ln \left(\frac{p}{1-p}\right)=\log i t(p)=\eta, \quad p=\frac{e^{\theta}}{1+e^{\theta}}, \quad \phi=a(\phi)=1, \\
b(\theta)=\ln \left(1+e^{\theta}\right), \quad b^{\prime}(\theta)=\frac{e^{\theta}}{1+e^{\theta}}=p=\mu, \\
b^{\prime \prime}(\theta)=\frac{e^{\theta}}{1+e^{\theta}}-\left(\frac{e^{\theta}}{1+e^{\theta}}\right)^{2}, \quad V(\mu)=\mu(1-\mu)
\end{array}
$$

co udowadnia, e rozkad ten naley do OEF.

Zdefiniujmy teraz log-likelihood:

$$
\ell(\beta \mid y)=\sum_{i=1}^{n}(y_{i} \ln(p)+(1-y_{i}) \ln(1-p))=\sum_{i=1}^{n}-\ln (1+e^{X_{i *} \beta})+\sum_{i=1}^{n} y_{i}(X_{i *} \beta).
$$
Aby obliczy wartoci parametrw $\beta$ naleaoby rozwiza rwnania wiarygodnoci:

$$
\frac{\partial \ell}{\partial \beta_{j}}=\sum_{i=1}^{n} x_{i j}\left(y_{i}-\frac{1}{1+e^{-\left(x_{i j} \beta_{j}\right)}}\right)=0
$$
Niestety w tym jak i w wielu innych przypadkach nie da si jawnie okreslic rozwiazania. Mozna je za to znale iteracyjnie korzystajc np z algorytmu **spadku wzdu gradientw** lub algorytmu **Newtona-Raphsona**:

$$
\beta^{(k+1)}=\beta^{(k)}-\frac{u\left(\beta^{(k)}\right)}{\mathscr{H}\left(\beta^{(k)}\right)}
$$

gdzie $u\left(\beta^{(k)}\right)$ oznacza gradient, a $\mathscr{H}\left(\beta^{(k)}\right)$ Hessian log-likelihood.

## Modelowanie

W naszymch danych znajduj si zmienne skorelowane i nieistotne, to co moglibymy zrobi to rcznie usuwa po jednej nieistotnej zmiennej a do momentu uzyskania modelu ze zmiennyi istotnymi. Niestety gdy odejmiemy/dodamy jakas zmienn istotnoci innych zmiennych mog si zmieni.

Jak wic automatycznie stworzy model ?

1. Mona sprawdzi wszystkie kombinacje zmiennych, ktrych jest $2^p$ - kiepski pomys przy duej liczbie zmiennych
2. Model regresji krokowej (dodajemy lub odejmujemy jedna zmienn na raz). Model wybierany jest pod wzgledem kryterium informacyjnego $AIC = -2\ell +2p$ - im mniejsze tym lepiej.
3. Uzy regularyzacji LASSO - kolejne zajcia.

Podzielmy teraz nasz oczyszczony zbir na treningowy i testowy, a nastepnie zestandaryzujmy zmienne

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
wdbc_small = wdbc[['Diagnosis', 'area', 'smoothness_Worst', 'texture', 'symmetry_Worst']]
train_data, test_data = train_test_split(wdbc_small, test_size=0.2, random_state=44)

y_train = train_data['Diagnosis']
X_train = train_data.drop('Diagnosis', axis=1)
y_test = test_data['Diagnosis']
X_test = test_data.drop('Diagnosis', axis=1)

scaler = StandardScaler()
X_train_norm = sm.add_constant(scaler.fit_transform(X_train))
X_test_norm = sm.add_constant(scaler.transform(X_test))
```

```python
logit_model = sm.Logit(y_train, X_train_norm).fit()
```

```python
logit_model.summary2()
```

Pozostaje jeszcze pytanie jak interpretowa wspczynniki modelu:

Wiemy, e 

$$
log(p/(1-p)) = betaX
$$

a wic

$$
p/(1-p) = e^{betaX}
$$

$p/(1-p)$ to tak zwana szansa (odds). Mwi ile razy bardziej prawdopodobne jest zdarzenie $p$ od $1-p$

```python
coefficients = logit_model.params
coefficients
```

Wyjciowe prawdopodobiestwo, przy wartosciach parametrw rwnych 0 wynosi:

```python
odds_0 = np.exp(coefficients[0])
print(odds_0)
probability_0 = sigmoid(coefficients[0])
print(probability_0)
```

Jak zmieni si szansa i prawdopodobiestwo gdy 'area' zwikszy si o 0.15 ?

```python
odds_1 = np.exp(coefficients[0] + 0.15*coefficients[1])
print(odds_1)
probability_1 = sigmoid(coefficients[0] + 0.15*coefficients[1])
print(probability_1)
```

Aby sprawdzic jaki wpyw ma zmienna area obliczamy iloraz szans (odds ratio):

$$
\frac{p_1/(1-p_1)}{p_0/(1-p_0)} = \frac{e^{\beta_0+0.15\beta_{area}}}{e^{\beta_0}} = e^{0.15*\beta_{area}}
$$

```python
print(np.exp(0.15*coefficients[1]))
```

Z powyszego wynika nastpujcy wniosek w modeli regresji logistycznej interpretacja zmiennych jest multiplikatywna: jeli warto zmiennej $x$ wzronie o $1$ to warto szansy wzronie $e^{\beta_x}$ razy.

Jak oceni model i dobra cutoff dla prawdopodobiestwa ?

```python
probs = logit_model.predict(X_test_norm)
pred_df = pd.DataFrame({'Diagnosis': y_test, 'Probability': probs})
sns.histplot(data=pred_df, x='Probability', hue='Diagnosis', element='step', stat='density', bins = 50)
```

Zacznijmy od cutoff rwnego 0.5

```python
pred_df["pred_cutoff_0.5"] = pred_df.Probability > 0.5
pivot_table = pred_df.drop(columns=["Probability"]).pivot_table(index='pred_cutoff_0.5', columns='Diagnosis', aggfunc=len, fill_value=0)
print(pivot_table)
```

Najprostsza miara do sprawdzenia to

$$
Accuracy = (TP + TN) / ALL
$$

 - TP - True Positives
 - TN - True Negatives
 - FP - False Positives (Type I error)
 - FN - False Negatives (Type II error)

```python
print("Accuracy for cutoff 0.5:", np.diag(pivot_table.values).sum() / pivot_table.values.sum())
```

Kiepska miara gdy mamy mocno niezbalansowane zbiory. Zamiast tego lepiej patrze na ponisze metryki:

$$
Precision = TP / (TP + FP) \\
Recall = TP / (TP + FN) \\
F_{\beta} = \frac{(1 + \beta^2) * (Precision * Recall)}{((\beta^2 * Precision) + Recall)}
$$

```python
precision = pivot_table.values[1,1] / np.sum(pivot_table.values[1,:])
print("Precision dla cutoffu 0.5:", precision)
recall = pivot_table.values[1,1] / np.sum(pivot_table.values[:,1])
print("Recall dla cutoffu 0.5:", recall)
f1 = (1 + 1**2) * precision * recall / (1**2 * precision + recall)
print("F1 dla cutoffu 0.5:", f1)
f2 = (1 + 2**2) * precision * recall / (2**2 * precision + recall)
print("F1 dla cutoffu 0.5:", f2)
```

Aby znale optymalny threshold (cutoff) zwyczajnie wybieramy jend z metyk i sprawdzamy jaki threshld j maksymalizuje:

```python
from sklearn.metrics import f1_score
thresholds = np.linspace(0, 1, num=100)
f1_scores = [f1_score(pred_df.Diagnosis, (pred_df.Probability >= threshold).astype(int)) for threshold in thresholds]
best_threshold = thresholds[np.argmax(f1_scores)]
print("Najlepszy threshold:", best_threshold)
pred_df["pred_cutoff_max_f1"] = pred_df.Probability > 0.5
pivot_table = pred_df.drop(columns=["Probability"]).pivot_table(index='pred_cutoff_max_f1', columns='Diagnosis', aggfunc=len, fill_value=0)
print(pivot_table)
```

Jeli chcemy spojrze na wyniki dla kadego cutoffu moemy si posuy Krzyw ROC (receiver operating characteristic curve) i AUC.

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(pred_df.Diagnosis, pred_df.Probability)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```
