```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

colnames = ["Age", "Workclass", "Final Weight", "Education", "Education Number of Years",
"Marital-status", "Occupation", "Relationship", "Race", "Sex", "Capital-gain", "Capital-loss",
"Hours-per-week", "Native-country", "Income"]

adult = pd.read_csv("data/adult-all.txt", names=colnames, index_col=False, na_values='?')
adult.head(5)
```

**Age**: ciągła wartość liczbowa, reprezentująca wiek osoby (np. 39).

**Workclass**: zmienna kategoryczna, sektor pracy.

**Final Weight**: waga próbki, czyli liczbę osób, które reprezentuje każda obserwacja w populacji docelowej.

**Education**: zmienna kategoryczna, opisująca poziom wykształcenia osoby.

**Education Number of Years**: liczba lat edukacji.

**Marital-status**: zmienna kategoryczna, opisująca stan cywilny osoby.

**Occupation**: zmienna kategoryczna, opisująca zawód osoby.

**Relationship**: zmienna kategoryczna, opisująca relacje rodzinne osoby.

**Race**: zmienna kategoryczna, opisująca przynależność etniczną osoby.

**Sex**: zmienna binarna, opisująca płeć osoby (mężczyzna lub kobieta).

**Capital-gain**: to zysk z inwestycji, który jest równy różnicy między ceną zakupu a ceną sprzedaży aktywów inwestycyjnych.

**Capital-loss**: to straty z inwestycji, które są również równoważne różnicy między ceną zakupu a ceną sprzedaży aktywów inwestycyjnych.

**Hours-per-week**: Liczba godzin przepracowanych w tygodniu.

**Native-country**: kraj pochodzenia.

**Income**: zmienna binarna, opisująca czy osoba ma dochód powyżej lub poniżej 50 tys. dolarów rocznie (">50K" lub "<=50K"). Jest to zmienna celu, którą chcemy przewidywać w zadaniu klasyfikacji.

Połączymy kilka kategorii dla wybranych zmiennych, wurzucimy skorelowane zmienne oraz outliery:

```python
adult.drop(['Education Number of Years'], axis = 1, inplace = True)
adult['Education'].replace(['11th', '9th', '7th-8th', '5th-6th', '10th', '1st-4th', 'Preschool', '12th'],
                             'School', inplace = True)

adult['Race'].replace(['Black', 'Asian-Pac-Islander', 'Amer-Indian-Eskimo'], 'Other', inplace = True)
adult['Marital-status'].replace(['Married-AF-spouse'], 'Married-civ-spouse', inplace = True)
adult['Native-country'] = ['United-States' if c == 'United-States' else 'Other' for c in adult['Native-country']]
adult = adult[adult['Capital-gain'] < 80000]
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from imblearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

X, y = adult.drop(columns=['Income'], axis=1), adult['Income']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

cat_ix = X.select_dtypes(include=['object', 'bool']).columns
num_ix = X.select_dtypes(include=['int64', 'float64']).columns
```

```python
names = ['CART', 'BAG', 'RF', 'GBM']
models = [
    DecisionTreeClassifier(),
    BaggingClassifier(n_estimators=100),
    RandomForestClassifier(n_estimators=100),
    GradientBoostingClassifier(n_estimators=100)
]
```

```python
results = list()
for i in range(len(names)):
    transformation_steps = [
        ('c', OneHotEncoder(handle_unknown='ignore'), cat_ix),
        ('n', MinMaxScaler(), num_ix)
    ]
    ct = ColumnTransformer(transformation_steps)
    pipeline = Pipeline(
        steps=[
            ('transform', ct),
            ('smote', SMOTE()),
            ('model', models[i])
        ])
    cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=0)
    scores = cross_val_score(pipeline, X_train, y_train, scoring='f1', cv=cv, n_jobs=-1)
    results.append(scores)
    print('>%s %.3f (%.3f)' % (names[i], np.mean(scores), np.std(scores)))
plt.boxplot(results, labels=names, showmeans=True)
plt.show()
```

```python
# >CART 0.602 (0.006)
# >BAG 0.647 (0.007)
# >RF 0.672 (0.006)
# >GBM 0.689 (0.005)
```

```python
transformation_steps = [
    ('c', OneHotEncoder(handle_unknown='ignore'), cat_ix),
    ('n', MinMaxScaler(), num_ix)
]
ct = ColumnTransformer(transformation_steps)
pipeline = Pipeline(
    steps=[
        ('transform', ct),
        ('smote', SMOTE()),
        ('model', GradientBoostingClassifier())
    ])
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=0)
param_grid = {
    "model__n_estimators": [50, 100]
}
grid_search = GridSearchCV(
    estimator=pipeline, param_grid=param_grid, cv=cv, scoring="f1_macro"
)
grid_search.fit(X_train, y_train)
```

```python
best_estimator = grid_search.best_estimator_

y_pred = best_estimator.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print('ROC AUC score:', roc_auc_score(y_test, y_pred))
```
